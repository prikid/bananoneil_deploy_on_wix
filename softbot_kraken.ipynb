```python
import requests
import json
import pandas as pd
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
#VARIABLES TO GET
#(<time>, <open>, <high>, <low>, <close>, <vwap>, <volume>, <count>) variables taken. Candlestick chart. Euro currency.
```

```python
#CONVERT UNIX TIME TO DAY/MONTH/YEAR STAMP
from datetime import datetime
import datetime as dt
import numpy as np
time_conv = np.vectorize(dt.datetime.fromtimestamp)
```

```python
#BITCOIN GET REQUEST OF DATA FROM KRAKEN'S API, ENCODING TO UTF-8 TO INTERPRETE, CLEANING DATA TO A SINGLES DATA FRAMES 
xbt = requests.get("https://api.kraken.com/0/public/OHLC?pair=XBTEUR&interval=1440")
xbt.encoding = "utf-8"
xbtjson = xbt.json()
b = xbtjson["result"]["XXBTZEUR"]

btime = []
for e in b:
  btime.append(e[0])

btime = time_conv(btime)
btime = pd.DataFrame(btime)

bopen = []
for e in b:
  bopen.append(e[1])

bopen = pd.DataFrame(bopen)

bhigh = []
for e in b:
  bhigh.append(e[2])

bhigh = pd.DataFrame(bhigh)

blow = []
for e in b:
  blow.append(e[3])

blow = pd.DataFrame(blow)

bclose = []
for e in b:
  bclose.append(e[4])

bclose = pd.DataFrame(bclose)

bvwap = []
for e in b:
  bvwap.append(e[5])

bvwap = pd.DataFrame(bvwap)

bvol = []
for e in b:
  bvol.append(e[6])

bvol = pd.DataFrame(bvol)
```

```python
#ETHERUM GET REQUEST OF DATA FROM KRAKEN'S API, ENCODING TO UTF-8 TO INTERPRETE, CLEANING DATA TO A SINGLES DATA FRAMES 
eth = requests.get("https://api.kraken.com/0/public/OHLC?pair=ETHEUR&interval=1440")
eth.encoding = "utf-8"
ethjson = eth.json()
e = ethjson["result"]["XETHZEUR"]

etime = []
for p in e:
  etime.append(p[0])

etime = time_conv(etime)
etime = pd.DataFrame(etime)

eopen = []
for p in e:
  eopen.append(p[1])

eopen = pd.DataFrame(eopen)

ehigh = []
for p in e:
  ehigh.append(p[2])

ehigh = pd.DataFrame(ehigh)

elow = []
for p in e:
  elow.append(p[3])

elow = pd.DataFrame(elow)

eclose = []
for p in e:
  eclose.append(p[4])

eclose = pd.DataFrame(eclose)

evwap = []
for p in e:
  evwap.append(p[5])

evwap = pd.DataFrame(evwap)

evol = []
for p in e:
  evol.append(p[6])

evol = pd.DataFrame(evol)
```

```python
#MONERO #GET REQUEST OF DATA FROM KRAKEN'S API, ENCODING TO UTF-8 TO INTERPRETE, CLEANING DATA TO A SINGLES DATA FRAMES 
xmr = requests.get("https://api.kraken.com/0/public/OHLC?pair=xmrEUR&interval=1440")
xmr.encoding = "utf-8"
xmrjson = xmr.json()
x = xmrjson["result"]["XXMRZEUR"]

xtime = []
for e in x:
  xtime.append(e[0])

xtime = time_conv(xtime)
xtime = pd.DataFrame(xtime)

xopen = []
for e in x:
  xopen.append(e[1])

xopen = pd.DataFrame(xopen)

xhigh = []
for e in x:
  xhigh.append(e[2])

xhigh = pd.DataFrame(xhigh)

xlow = []
for e in x:
  xlow.append(e[3])

xlow = pd.DataFrame(xlow)

xclose = []
for e in x:
  xclose.append(e[4])

xclose = pd.DataFrame(xclose)

xvwap = []
for e in x:
  xvwap.append(e[5])

xvwap = pd.DataFrame(xvwap)

xvol = []
for e in x:
  xvol.append(e[6])

xvol = pd.DataFrame(xvol)
```

```python
#CONCATENATE DATAFRAMES
```

```python
#BITCOIN DATAFRAME

#name to columns
btc_dataframe = pd.concat([btime,bopen,bhigh,blow,bclose,bvwap,bvol], axis=1)
btc_cols = ["time","open","high","low","close","volweightprice","volume"]
btc_dataframe.columns = btc_cols

#string to floats
btc_dataframe.open = btc_dataframe.open.astype(float)
btc_dataframe.high = btc_dataframe.high.astype(float)
btc_dataframe.low = btc_dataframe.low.astype(float)
btc_dataframe.close = btc_dataframe.close.astype(float)
btc_dataframe.volweightprice = btc_dataframe.volweightprice.astype(float)
btc_dataframe.volume = btc_dataframe.volume.astype(float)
btc_dataframe.head()
```

```python
#ETHERUM DATAFRAME

#name to columns
eth_dataframe = pd.concat([etime,eopen,ehigh,elow,eclose,evwap,evol], axis=1)
eth_cols = ["time","open","high","low","close","volweightprice","volume"]
eth_dataframe.columns = eth_cols

#string to floats
eth_dataframe.open = eth_dataframe.open.astype(float)
eth_dataframe.high = eth_dataframe.high.astype(float)
eth_dataframe.low = eth_dataframe.low.astype(float)
eth_dataframe.close = eth_dataframe.close.astype(float)
eth_dataframe.volweightprice = eth_dataframe.volweightprice.astype(float)
eth_dataframe.volume = eth_dataframe.volume.astype(float)
eth_dataframe.head()
```

```python
#MONERO DATAFRAME

#name to columns
xmr_dataframe = pd.concat([xtime,xopen,xhigh,xlow,xclose,xvwap,xvol], axis=1)
xmr_cols = ["time","open","high","low","close","volweightprice","volume"]
xmr_dataframe.columns = xmr_cols

#string to floats
xmr_dataframe.open = xmr_dataframe.open.astype(float)
xmr_dataframe.high = xmr_dataframe.high.astype(float)
xmr_dataframe.low = xmr_dataframe.low.astype(float)
xmr_dataframe.close = xmr_dataframe.close.astype(float)
xmr_dataframe.volweightprice = xmr_dataframe.volweightprice.astype(float)
xmr_dataframe.volume = xmr_dataframe.volume.astype(float)
xmr_dataframe.head()
```

```python
#pd dataframe close price for 1 day#
bclose1= pd.DataFrame(btc_dataframe["close"])
eclose1= pd.DataFrame(eth_dataframe["close"])
xclose1= pd.DataFrame(xmr_dataframe["close"])

#pd dataframe for 7 days#
bclose7= pd.DataFrame(btc_dataframe["close"])
eclose7= pd.DataFrame(eth_dataframe["close"])
xclose7= pd.DataFrame(xmr_dataframe["close"])

#pd dataframe for 30 days#
bclose30= pd.DataFrame(btc_dataframe["close"])
eclose30= pd.DataFrame(eth_dataframe["close"])
xclose30= pd.DataFrame(xmr_dataframe["close"])
```

```python
#Days to predict 
future_days_1 = 1
future_days_7 = 7
future_days_30 = 30
```

```python
#Add, column "prediction" (prediction = close - days to predict)
bclose1["Prediction"] = bclose1.shift(-future_days_1)
bclose7["Prediction"] = bclose7.shift(-future_days_7)
bclose30["Prediction"] = bclose30.shift(-future_days_30)

eclose1["Prediction"] = eclose1.shift(-future_days_1)
eclose7["Prediction"] = eclose7.shift(-future_days_7)
eclose30["Prediction"] = eclose30.shift(-future_days_30)

xclose1["Prediction"] = xclose1.shift(-future_days_1)
xclose7["Prediction"] = xclose7.shift(-future_days_7)
xclose30["Prediction"] = xclose30.shift(-future_days_30)
```

```python
#numpay array X from "close", 1 dimmension, - days to predict
bxarray1 = np.array(bclose1.drop(["Prediction"], 1))[:-future_days_1]
bxarray7 = np.array(bclose7.drop(["Prediction"], 1))[:-future_days_7]
bxarray30 = np.array(bclose30.drop(["Prediction"], 1))[:-future_days_30]

exarray1 = np.array(eclose1.drop(["Prediction"], 1))[:-future_days_1]
exarray7 = np.array(eclose7.drop(["Prediction"], 1))[:-future_days_7]
exarray30 = np.array(eclose30.drop(["Prediction"], 1))[:-future_days_30]

xxarray1 = np.array(xclose1.drop(["Prediction"], 1))[:-future_days_1]
xxarray7 = np.array(xclose7.drop(["Prediction"], 1))[:-future_days_7]
xxarray30 = np.array(xclose30.drop(["Prediction"], 1))[:-future_days_30]
```

```python
#numpy array Y from prediction'2
byarray1 = np.array(bclose1["Prediction"])[:-future_days_1]
byarray7 = np.array(bclose7["Prediction"])[:-future_days_7]
byarray30 = np.array(bclose30["Prediction"])[:-future_days_30]

eyarray1 = np.array(eclose1["Prediction"])[:-future_days_1]
eyarray7 = np.array(eclose7["Prediction"])[:-future_days_7]
eyarray30 = np.array(eclose30["Prediction"])[:-future_days_30]

xyarray1 = np.array(xclose1["Prediction"])[:-future_days_1]
xyarray7 = np.array(xclose7["Prediction"])[:-future_days_7]
xyarray30 = np.array(xclose30["Prediction"])[:-future_days_30]
```

```python
#TRAIN AND TEST#
from sklearn.model_selection import train_test_split

#Bitcoin
xb_train1, xb_test1, yb_train1, yb_test1 = train_test_split(bxarray1, byarray1, test_size = 0.2)
xb_train7, xb_test7, yb_train7, yb_test7 = train_test_split(bxarray7, byarray7, test_size = 0.2)
xb_train30, xb_test30, yb_train30, yb_test30 = train_test_split(bxarray30, byarray30, test_size = 0.2)

#Etherum 
xe_train1, xe_test1, ye_train1, ye_test1 = train_test_split(exarray1, eyarray1, test_size = 0.2)
xe_train7, xe_test7, ye_train7, ye_test7 = train_test_split(exarray7, eyarray7, test_size = 0.2)
xe_train30, xe_test30, ye_train30, ye_test30 = train_test_split(exarray30, eyarray30, test_size = 0.2)

#Monero
xx_train1, xx_test1, yx_train1, yx_test1 = train_test_split(xxarray1, xyarray1, test_size = 0.2)
xx_train7, xx_test7, yx_train7, yx_test7 = train_test_split(xxarray7, xyarray7, test_size = 0.2)
xx_train30, xx_test30, yx_train30, yx_test30 = train_test_split(xxarray30, xyarray30, test_size = 0.2)
```

```python
#IMPORT MODELS#
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
```

```python
#IMPLEMENTING MODELS#

#DECISION TREE REGRESSOR#
treeb1 = DecisionTreeRegressor().fit(xb_train1, yb_train1)
treeb7 = DecisionTreeRegressor().fit(xb_train7, yb_train7)
treeb30 = DecisionTreeRegressor().fit(xb_train30, yb_train30)

treee1 = DecisionTreeRegressor().fit(xe_train1, ye_train1)
treee7 = DecisionTreeRegressor().fit(xe_train7, ye_train7)
treee30 = DecisionTreeRegressor().fit(xe_train30 , ye_train30)

treex1 = DecisionTreeRegressor().fit(xx_train1, yx_train1)
treex7 = DecisionTreeRegressor().fit(xx_train7, yx_train7)
treex30 = DecisionTreeRegressor().fit(xx_train30, yx_train30)

#LINEAR REGRESSION#
lrb1 = LinearRegression().fit(xb_train1, yb_train1)
lrb7 = LinearRegression().fit(xb_train7, yb_train7)
lrb30 = LinearRegression().fit(xb_train30, yb_train30)

lre1 = LinearRegression().fit(xe_train1, ye_train1)
lre7 = LinearRegression().fit(xe_train7, ye_train7)
lre30 = LinearRegression().fit(xe_train30, ye_train30)

lrx1 = LinearRegression().fit(xx_train1, yx_train1)
lrx7 = LinearRegression().fit(xx_train7, yx_train7)
lrx30 = LinearRegression().fit(xx_train30, yx_train30)
```

```python
#ACURACY  DECISION TREE # 
print(treeb1.score(xb_test1, yb_test1))
print(treeb7.score(xb_test7, yb_test7))
print(treeb30.score(xb_test30, yb_test30))

print(treee1.score(xe_test1, ye_test1))
print(treee7.score(xe_test7, ye_test7))
print(treee30.score(xe_test30, ye_test30))

print(treex1.score(xx_test1, yx_test1))
print(treex7.score(xx_test7, yx_test7))
print(treex30.score(xx_test30, yx_test30))
```

```python
#ACURACY LINEAR REGRESSION#
print(lrb1.score(xb_test1, yb_test1))
print(lrb7.score(xb_test7, yb_test7))
print(lrb30.score(xb_test30, yb_test30))

print(lre1.score(xe_test1, ye_test1))
print(lre7.score(xe_test7, ye_test7))
print(lre30.score(xe_test30, ye_test30))

print(lrx1.score(xx_test1, yx_test1))
print(lrx7.score(xx_test7, yx_test7))
print(lrx30.score(xx_test30, yx_test30))
```

```python
#X FUTURE BITCOIN 1, 7, 30 #
b_future1 = bclose1.drop(["Prediction"], 1)[:-future_days_1]
b_future1 = b_future1.tail(future_days_1)
b_future1 = np.array(b_future1)

b_future7 = bclose7.drop(["Prediction"], 1)[:-future_days_7]
b_future7 = b_future7.tail(future_days_7)
b_future7 = np.array(b_future7)


b_future30 = bclose30.drop(["Prediction"], 1)[:-future_days_30]
b_future30 = b_future30.tail(future_days_30)
b_future30 = np.array(b_future30)

#X FUTURE ETHERUM 1, 7, 30 #
e_future1 = eclose1.drop(["Prediction"], 1)[:-future_days_1]
e_future1 = e_future1.tail(future_days_1)
e_future1 = np.array(e_future1)

e_future7 = eclose7.drop(["Prediction"], 1)[:-future_days_7]
e_future7 = e_future7.tail(future_days_7)
e_future7 = np.array(e_future7)


e_future30 = eclose30.drop(["Prediction"], 1)[:-future_days_30]
e_future30 = e_future30.tail(future_days_30)
e_future30 = np.array(e_future30)

#X FUTURE MONERO 1, 7, 30#

x_future1 = xclose1.drop(["Prediction"], 1)[:-future_days_1]
x_future1 = x_future1.tail(future_days_1)
x_future1 = np.array(x_future1)

x_future7 = xclose7.drop(["Prediction"], 1)[:-future_days_7]
x_future7 = x_future7.tail(future_days_7)
x_future7 = np.array(x_future7)


x_future30 = xclose30.drop(["Prediction"], 1)[:-future_days_30]
x_future30 = x_future30.tail(future_days_30)
x_future30 = np.array(x_future30)
```
